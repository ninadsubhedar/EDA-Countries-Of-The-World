Defining common function to plot Country vs GDP bar plot

```python
def plotCountryVsGDP(dataframe):
    fig, ax = plt.subplots(figsize=(16,6))
    sns.barplot(x='Country', y='GDP ($ per capita)', data=dataframe, palette='Set1')
    ax.set_xlabel(ax.get_xlabel(), labelpad=15)
    ax.set_ylabel(ax.get_ylabel(), labelpad=30)
    ax.xaxis.label.set_fontsize(16)
    ax.yaxis.label.set_fontsize(16)
    plt.xticks(rotation=90)
    plt.show()
```

## Multivariate Analysis
#### Understanding Correlation between GDP and other fields in the dataset

```python
gdpSeries = df['GDP ($ per capita)']
df.corrwith(gdpSeries).round(2)
```

##### Visualizing the correlation using a heat map
The red boxes indicate high positive correlation<br>
And the blue boxes indicate high negative correlation<br>
The color intensity indicates the level of correlation.<br> 
Darker shade indicates high and lighter shades indicate low<br>
The correlation is based on Pearson's r value calculated for each correlation 

```python
def plotHeatMap(dataFrame,vmax):
    """
    Plots a heatmap of correlation between data.
    
    Args:
        dataFrame: A pandas data frame for which the heatmap needs to be plotted.
    
    Return:
        This function returns nothing but plots a heatmap.
    """
    
    plt.figure(figsize=(16,12))
    corr = dataFrame.corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    with sns.axes_style("white"):
        ax = sns.heatmap(data=corr,annot=True,mask=mask,fmt='.2f',vmax=vmax,cmap='coolwarm',square=True)
    plt.xticks(rotation=90)
    plt.show()
    
plotHeatMap(df.iloc[:,2:],0.8)
```

##### Observation 5
1. `Literacy`,`Phones (per 1000)`,`Infant mortality (per 1000 births)`,`Birthrate`,`Deathrate` have highest absolute value of  correlation with `GDP ($ per capita)`

We can zoom in on the correlation for the above fields and visualize them as shown in the below heat map

```python
# choose attributes which shows relation
x = df[['GDP ($ per capita)','Literacy (%)','Phones (per 1000)','Service','Infant mortality (per 1000 births)','Birthrate']]
# show corr of the same
plotHeatMap(x,0.9)
```

##### Visializing the tendancies for correlations with GDP

Here we visualize the tendancy of GDP with respect to other independant variables

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20,20))
plt.subplots_adjust(hspace=0.4)

corr_to_gdp = pd.Series()
for col in df.columns.values[2:]:
    if ((col!='GDP ($ per capita)')&(col!='Climate')):
        corr_to_gdp[col] = df['GDP ($ per capita)'].corr(df[col])
abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)
corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]


for i in range(4):
    for j in range(4):
        sns.regplot(x=corr_to_gdp.index.values[i*4+j], y='GDP ($ per capita)', data=df,
                   ax=axes[i,j], marker='.', order=2)
        title = 'correlation='+str(corr_to_gdp[i*4+j])
        axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
plt.show()
```

##### Observation 6
1. The field `Phones (per 1000)` has a strong positive linear relation with the `GDP ($ per capita)` field.
2. `Service`, `Birthrate`, `Infant mortality (per 1000 births)`, `Agriculture`, `Literacy (%)` has a non linear relation.

The linear relations van be seen in the below pair plots for `GDP ($ per capita)`, `Service`, `Phones (per 1000)`
